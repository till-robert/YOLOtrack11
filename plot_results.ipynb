```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
# my_script.py

```

```python
def plot_results(fname, start_epoch=0, log=True):
    plt.figure(figsize=(12,5)) if not plt.get_fignums() else plt.clf()
    res = pd.read_csv(fname)
    for r in res:
    
        if "val" in r:
            plt.subplot(1,2,1)
            plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r)
            # print(res["epoch"][start_epoch:],res[r][start_epoch:])
        elif "MSE" in r:
            plt.subplot(1,2,2)
            plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r.replace("metrics/",""))
        elif "metric" in r:
            plt.subplot(1,2,2)
            if log:
                plt.plot(res["epoch"][start_epoch:],1-res[r][start_epoch:], label="1-"+r.replace("metrics/",""))
            else:
                plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r.replace("metrics/",""))
    plt.subplot(1,2,1)
    plt.xlabel("Epoch")
    plt.ylabel("loss")
    if log:
        plt.yscale("log")
    plt.legend()
    plt.subplot(1,2,2)
    plt.ylabel("metric")
    plt.xlabel("Epoch")
    if log:
        plt.yscale("log")
    plt.legend()
    plt.show()
```

```python
plot_results("../ultralytics/runs/zaxis/train163/results.csv")
```

```python
plot_results("runs/zaxis/train113/results.csv",log=False)
```
