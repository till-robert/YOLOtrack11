```python
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib widget
# import seaborn as sns
# sns.set_theme()
# my_script.py'

```

```python
def plot_results(fname, start_epoch=0, log=True, show_training_loss=True):
    plt.figure(figsize=(12,5)) if not plt.get_fignums() else plt.clf()
    res = pd.read_csv(fname)
    for r in res:
    
        if "val" in r:
            plt.subplot(1,2,1)
            plt.title("dashed: train loss, solid: validation loss")
            l,=plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r.replace("val/",""))
            if show_training_loss:
                train_loss_key = r.replace("val","train")
                plt.plot(res["epoch"][start_epoch:],res[train_loss_key][start_epoch:], ls="dashed", color=l.get_color())

            # print(res["epoch"][start_epoch:],res[r][start_epoch:])
        elif "rms." in r or "MSE" in r:
            plt.subplot(1,2,2)
            plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r.replace("metrics/",""))
        elif "metric" in r:
            plt.subplot(1,2,2)
            if log:
                plt.plot(res["epoch"][start_epoch:],1-res[r][start_epoch:], label="1-"+r.replace("metrics/",""))
            else:
                plt.plot(res["epoch"][start_epoch:],res[r][start_epoch:], label=r.replace("metrics/",""))
    plt.subplot(1,2,1)
    plt.xlabel("Epoch")
    plt.ylabel("loss")
    if log:
        plt.yscale("log")
    plt.legend()
    plt.subplot(1,2,2)
    plt.ylabel("metric")
    plt.xlabel("Epoch")
    if log:
        plt.yscale("log")
    plt.legend()
    plt.show()
    return plt.gcf()
```

```python
plot_results("../ultralytics/runs/pose/train64/results.csv", log=False)

```
